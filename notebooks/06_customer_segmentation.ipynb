```python
# 06_customer_segmentation.ipynb
# Banking Customer Segmentation Analysis

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from datetime import datetime
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.cm as cm
from matplotlib.colors import Normalize
import joblib

# For 3D visualization
from mpl_toolkits.mplot3d import Axes3D

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')

# Define plot aesthetics
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# 1. Load the data with CLV predictions
print("Loading customer data with CLV predictions...")

# Define paths
data_dir = '../data/processed/'
models_dir = '../models/'
output_dir = data_dir
os.makedirs(output_dir, exist_ok=True)

# Load customer features with CLV
try:
    features = pd.read_csv(f'{data_dir}customer_features_with_clv.csv')
    print(f"Loaded {len(features)} customer records with {features.shape[1]} features")
except FileNotFoundError:
    print("Error: Could not find customer_features_with_clv.csv file.")
    print("Please run the CLV modeling notebook first to generate this file.")
    # If needed, you could add a fallback to load from banking_customer_features.csv here
    raise

# Display basic information about the dataset
print("\nDataset Overview:")
print(f"Number of customers: {len(features)}")
print(f"Number of features: {features.shape[1]}")
print(f"Feature columns: {', '.join(features.columns[:10])}...")

# Display the first few rows
display(features.head())

# 2. Feature selection for segmentation
print("\nSelecting features for segmentation...")

# Check for the presence of expected features
expected_features = [
    'recency_days', 'frequency', 'monetary_value', 
    'balance_range', 'avg_transaction_amount',
    'product_diversity_score', 'loan_count', 'card_count', 
    'permanent_order_count', 'avg_account_balance',
    'clv_1yr', 'predicted_transactions_1yr'
]

# Check which expected features are available
available_features = [f for f in expected_features if f in features.columns]
missing_features = [f for f in expected_features if f not in features.columns]

if missing_features:
    print(f"Warning: Some expected features are missing: {missing_features}")
    
    # Check for equivalent features with different names
    if 'product_diversity_score' not in features.columns and 'product_diversity' in features.columns:
        print("Found 'product_diversity' as alternative to 'product_diversity_score'")
        available_features.append('product_diversity')
        
    if 'clv_1yr' not in features.columns and 'customer_lifetime_value' in features.columns:
        print("Found 'customer_lifetime_value' as alternative to 'clv_1yr'")
        available_features.append('customer_lifetime_value')

# Define groupings of features for more organized selection
feature_groups = {
    'RFM Features': ['recency_days', 'frequency', 'monetary_value'],
    'Transaction Features': ['avg_transaction_amount', 'transaction_amount_volatility', 
                            'balance_range', 'monetary_per_transaction'],
    'Product Features': ['product_diversity_score', 'product_diversity', 'loan_count', 
                         'card_count', 'permanent_order_count'],
    'Balance Features': ['avg_balance', 'min_balance', 'max_balance', 'balance_volatility'],
    'CLV Features': ['clv_1yr', 'customer_lifetime_value', 'predicted_transactions_1yr']
}

# Select features for segmentation, filtering out unavailable ones
segmentation_features = []
feature_group_status = {}

for group, group_features in feature_groups.items():
    # Get features from this group that are available in the dataset
    available_group_features = [f for f in group_features if f in features.columns]
    feature_group_status[group] = available_group_features
    segmentation_features.extend(available_group_features)

# Ensure we have no duplicates
segmentation_features = list(dict.fromkeys(segmentation_features))

print("\nSelected feature groups for segmentation:")
for group, selected_features in feature_group_status.items():
    print(f"- {group}: {', '.join(selected_features) if selected_features else 'None available'}")

# Create the final feature subset for segmentation
segment_features = features[segmentation_features].copy()

# Handle missing values
missing_values = segment_features.isnull().sum()
if missing_values.sum() > 0:
    print(f"\nFound {missing_values.sum()} missing values in the selected features")
    print(missing_values[missing_values > 0])
    print("Filling missing values with zeros")
    segment_features = segment_features.fillna(0)

# 3. Feature scaling and data preparation
print("\nStandardizing features for segmentation...")

# Create a scaled version of the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(segment_features)
scaled_df = pd.DataFrame(scaled_features, columns=segment_features.columns)

# Save the scaler for future use
joblib.dump(scaler, f'{models_dir}segmentation_scaler.pkl')

# Display summary statistics of scaled features
display(scaled_df.describe().T)

# Check for any remaining issues after scaling
print(f"NaN values after scaling: {scaled_df.isnull().sum().sum()}")
print(f"Infinite values after scaling: {np.isinf(scaled_features).sum()}")

# Replace any NaN or infinite values that might have been introduced during scaling
if scaled_df.isnull().sum().sum() > 0 or np.isinf(scaled_features).sum() > 0:
    print("Replacing NaN and infinite values with zeros")
    scaled_df = scaled_df.replace([np.inf, -np.inf], np.nan)
    scaled_df = scaled_df.fillna(0)
    scaled_features = scaled_df.values

# 4. Determine optimal number of clusters
print("\nDetermining optimal number of clusters...")

# Initialize metrics lists
wcss = []  # Within-Cluster Sum of Squares (Elbow method)
silhouette_scores = []  # Silhouette Score
ch_scores = []  # Calinski-Harabasz Index
db_scores = []  # Davies-Bouldin Index

# Set range of clusters to evaluate
min_clusters = 2
max_clusters = 10
cluster_range = range(min_clusters, max_clusters + 1)

# Compute clustering metrics for different numbers of clusters
for n_clusters in cluster_range:
    print(f"Testing {n_clusters} clusters...", end='\r')
    
    # Initialize and fit KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(scaled_features)
    
    # Calculate metrics
    wcss.append(kmeans.inertia_)
    
    if len(np.unique(cluster_labels)) > 1:  # Ensure we have at least 2 clusters for validity metrics
        silhouette_scores.append(silhouette_score(scaled_features, cluster_labels))
        ch_scores.append(calinski_harabasz_score(scaled_features, cluster_labels))
        db_scores.append(davies_bouldin_score(scaled_features, cluster_labels))
    else:
        silhouette_scores.append(0)
        ch_scores.append(0)
        db_scores.append(float('inf'))

print("\nCluster metrics computed successfully")

# Plot the clustering metric results
plt.figure(figsize=(14, 10))

# Elbow Method - Lower is better, look for the "elbow" point
plt.subplot(2, 2, 1)
plt.plot(cluster_range, wcss, marker='o', linestyle='-')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.grid(True)
plt.xticks(cluster_range)

# Silhouette Score - Higher is better (max 1)
plt.subplot(2, 2, 2)
plt.plot(cluster_range, silhouette_scores, marker='o', linestyle='-')
plt.title('Silhouette Score (Higher is Better)')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.xticks(cluster_range)

# Calinski-Harabasz Index - Higher is better
plt.subplot(2, 2, 3)
plt.plot(cluster_range, ch_scores, marker='o', linestyle='-')
plt.title('Calinski-Harabasz Index (Higher is Better)')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.grid(True)
plt.xticks(cluster_range)

# Davies-Bouldin Index - Lower is better
plt.subplot(2, 2, 4)
plt.plot(cluster_range, db_scores, marker='o', linestyle='-')
plt.title('Davies-Bouldin Index (Lower is Better)')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.grid(True)
plt.xticks(cluster_range)

plt.tight_layout()
plt.show()

# Organize the metrics in a DataFrame for easier analysis
metrics_df = pd.DataFrame({
    'n_clusters': list(cluster_range),
    'wcss': wcss,
    'silhouette_score': silhouette_scores,
    'calinski_harabasz_score': ch_scores,
    'davies_bouldin_score': db_scores
})

# Normalize metrics for easier comparison (0-1 scale where higher is better)
metrics_df['norm_wcss'] = 1 - (metrics_df['wcss'] - metrics_df['wcss'].min()) / (metrics_df['wcss'].max() - metrics_df['wcss'].min())
metrics_df['norm_silhouette'] = (metrics_df['silhouette_score'] - metrics_df['silhouette_score'].min()) / (metrics_df['silhouette_score'].max() - metrics_df['silhouette_score'].min())
metrics_df['norm_ch'] = (metrics_df['calinski_harabasz_score'] - metrics_df['calinski_harabasz_score'].min()) / (metrics_df['calinski_harabasz_score'].max() - metrics_df['calinski_harabasz_score'].min())
metrics_df['norm_db'] = 1 - (metrics_df['davies_bouldin_score'] - metrics_df['davies_bouldin_score'].min()) / (metrics_df['davies_bouldin_score'].max() - metrics_df['davies_bouldin_score'].min())

# Calculate an aggregate score (mean of normalized metrics)
metrics_df['aggregate_score'] = metrics_df[['norm_wcss', 'norm_silhouette', 'norm_ch', 'norm_db']].mean(axis=1)

# Display the metrics
print("\nClustering Metrics by Number of Clusters:")
display(metrics_df)

# Find the optimal number of clusters based on aggregate score
optimal_clusters = metrics_df.loc[metrics_df['aggregate_score'].idxmax(), 'n_clusters']
print(f"\nBased on aggregate score, the optimal number of clusters is: {optimal_clusters}")

# Alternative recommendation based on silhouette score
silhouette_optimal = metrics_df.loc[metrics_df['silhouette_score'].idxmax(), 'n_clusters']
if silhouette_optimal != optimal_clusters:
    print(f"Based on silhouette score alone, the optimal number would be: {silhouette_optimal}")

# Allow manual override - important for domain-specific considerations
# In banking, 5 clusters is often a practical number for segmentation strategies
manual_override = 5  # Set this to None if you don't want to override
if manual_override is not None:
    print(f"\nManually overriding to {manual_override} clusters based on business considerations")
    final_clusters = manual_override
else:
    final_clusters = optimal_clusters

print(f"\nProceeding with {final_clusters} clusters for customer segmentation")

# 5. Perform the final clustering
print("\nPerforming final clustering...")

# Initialize and fit the KMeans model with the chosen number of clusters
kmeans = KMeans(n_clusters=final_clusters, random_state=42, n_init=10, max_iter=300)
cluster_labels = kmeans.fit_predict(scaled_features)

# Add cluster labels to original features dataframe
features['cluster'] = cluster_labels

# Save the clustering model for future use
joblib.dump(kmeans, f'{models_dir}kmeans_model.pkl')

# 6. Analyze cluster characteristics
print("\nAnalyzing cluster characteristics...")

# Get cluster sizes
cluster_sizes = features['cluster'].value_counts().sort_index()
print("\nCluster Sizes:")
for cluster, size in cluster_sizes.items():
    print(f"Cluster {cluster}: {size} customers ({size/len(features)*100:.1f}%)")

# Calculate cluster centers (in original feature space)
cluster_centers_scaled = kmeans.cluster_centers_
cluster_centers_original = scaler.inverse_transform(cluster_centers_scaled)
centers_df = pd.DataFrame(cluster_centers_original, columns=segment_features.columns)

# Detailed cluster profile - compute mean values of key features per cluster
# Include more features than those used for clustering to get a comprehensive view
profile_features = [
    'recency_days', 'frequency', 'monetary_value', 
    'avg_transaction_amount', 'avg_balance', 'balance_range',
    'loan_count', 'card_count', 'permanent_order_count',
    'predicted_transactions_1yr', 'expected_avg_profit',
    'clv_1yr', 'customer_lifetime_value'
]

# Filter to only include available features
profile_features = [f for f in profile_features if f in features.columns]

# Calculate cluster profiles
cluster_profile = features.groupby('cluster')[profile_features].mean()

# Add count and percentage
cluster_counts = features['cluster'].value_counts().sort_index()
cluster_profile['count'] = cluster_counts
cluster_profile['percentage'] = (cluster_counts / len(features) * 100).round(1)

# Rank clusters by CLV (use available CLV column)
clv_col = 'clv_1yr' if 'clv_1yr' in features.columns else 'customer_lifetime_value'
if clv_col in features.columns:
    cluster_profile['value_rank'] = cluster_profile[clv_col].rank(ascending=False).astype(int)
    
    # Calculate total value by cluster
    cluster_total_value = features.groupby('cluster')[clv_col].sum()
    cluster_profile['total_value'] = cluster_total_value
    cluster_profile['value_percentage'] = (cluster_total_value / cluster_total_value.sum() * 100).round(1)

# Display cluster profiles
print("\nCluster Profiles (Mean Values):")
display(cluster_profile)

# 7. Create descriptive cluster names
# This is based on the cluster characteristics
# Banking-specific names that highlight key customer behavior patterns
print("\nAssigning descriptive names to clusters...")

# Initialize a dictionary to hold cluster names
cluster_names = {}

# Identify the key differentiating factors for each cluster
# We'll look at recency, frequency, monetary values, and product holdings
if clv_col in features.columns:
    # Sort clusters by value to assign consistent names from highest to lowest value
    sorted_clusters = cluster_profile.sort_values(clv_col, ascending=False).index
    
    # Prepare to assign names based on distinct characteristics
    for i, cluster in enumerate(sorted_clusters):
        profile = cluster_profile.loc[cluster]
        
        # Initialize name components
        value_tier = ""
        activity_level = ""
        product_focus = ""
        
        # Determine value tier based on CLV rank
        if i == 0:
            value_tier = "Premium"
        elif i == len(sorted_clusters) - 1:
            value_tier = "Basic"
        else:
            value_tier = "Standard"
            
        # Determine activity level based on frequency and recency
        # Higher frequency and lower recency means more active
        if profile['frequency'] > cluster_profile['frequency'].median():
            if profile['recency_days'] < cluster_profile['recency_days'].median():
                activity_level = "Active"
            else:
                activity_level = "Intermittent"
        else:
            if profile['recency_days'] > cluster_profile['recency_days'].median():
                activity_level = "Dormant"
            else:
                activity_level = "Low-Activity"
                
        # Determine product focus based on product holdings
        if 'loan_count' in profile and 'card_count' in profile and 'permanent_order_count' in profile:
            if profile['loan_count'] > cluster_profile['loan_count'].median():
                product_focus = "Borrowers"
            elif profile['card_count'] > cluster_profile['card_count'].median():
                product_focus = "Transactors"
            elif profile['permanent_order_count'] > cluster_profile['permanent_order_count'].median():
                product_focus = "Automated Users"
            elif profile['avg_balance'] > cluster_profile['avg_balance'].median():
                product_focus = "Savers"
            else:
                product_focus = "Basic Users"
        else:
            # Simplified names if specific product columns aren't available
            if profile['monetary_value'] > cluster_profile['monetary_value'].median():
                product_focus = "High-Value"
            else:
                product_focus = "Standard"
                
        # Combine components into a meaningful name
        cluster_names[cluster] = f"{value_tier} {activity_level} {product_focus}"
else:
    # Simplified naming if CLV isn't available
    for cluster in range(final_clusters):
        cluster_names[cluster] = f"Segment {cluster+1}"

# Display the cluster names
print("\nCluster Names:")
for cluster, name in cluster_names.items():
    print(f"Cluster {cluster}: {name}")

# Add names to the cluster profile
cluster_profile['segment_name'] = [cluster_names[i] for i in cluster_profile.index]

# 8. Visualize clusters
print("\nVisualizing customer segments...")

# Dimensionality reduction for visualization using PCA
pca = PCA(n_components=3)  # Reduce to 3 components for 3D visualization
principal_components = pca.fit_transform(scaled_features)

# Create a DataFrame with principal components
pca_df = pd.DataFrame(
    data=principal_components,
    columns=['Principal Component 1', 'Principal Component 2', 'Principal Component 3']
)
pca_df['Cluster'] = features['cluster']
pca_df['Cluster Name'] = pca_df['Cluster'].map(cluster_names)

# Check explained variance to understand how much information is retained
explained_variance = pca.explained_variance_ratio_
print(f"\nExplained variance by principal components: {explained_variance}")
print(f"Total variance explained by 3 components: {sum(explained_variance)*100:.2f}%")

# Create feature importance visualization for PCA components
pca_loadings = pca.components_
pca_feature_importance = pd.DataFrame(
    pca_loadings.T,
    columns=[f'PC{i+1}' for i in range(3)],
    index=segment_features.columns
)

plt.figure(figsize=(14, 8))
sns.heatmap(pca_feature_importance, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Importance in Principal Components')
plt.tight_layout()
plt.show()

# Create 2D PCA plot
plt.figure(figsize=(12, 10))
scatter = sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='Cluster Name',
    data=pca_df,
    palette='viridis',
    alpha=0.7,
    s=100
)

# Add cluster centroids to the plot
centroids_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(
    centroids_pca[:, 0],
    centroids_pca[:, 1],
    s=200,
    marker='X',
    c='red',
    edgecolor='black',
    label='Centroids'
)

# Add annotations for cluster names
for i, centroid in enumerate(centroids_pca):
    plt.annotate(
        cluster_names[i],
        (centroid[0], centroid[1]),
        textcoords="offset points",
        xytext=(0, 10),
        ha='center',
        fontsize=12,
        weight='bold'
    )

plt.title('Customer Segments Visualization (PCA)', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=14)
plt.ylabel('Principal Component 2', fontsize=14)
plt.legend(title='Segment', title_fontsize=12)
plt.tight_layout()
plt.show()

# Create 3D PCA plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# List of colors for clusters
colors = cm.viridis(np.linspace(0, 1, final_clusters))

# Plot each cluster
for cluster_id in range(final_clusters):
    cluster_data = pca_df[pca_df['Cluster'] == cluster_id]
    ax.scatter(
        cluster_data['Principal Component 1'],
        cluster_data['Principal Component 2'],
        cluster_data['Principal Component 3'],
        label=cluster_names[cluster_id],
        color=colors[cluster_id],
        alpha=0.7,
        s=50
    )

# Plot the centroids
ax.scatter(
    centroids_pca[:, 0],
    centroids_pca[:, 1],
    centroids_pca[:, 2],
    s=200,
    marker='X',
    c='red',
    edgecolor='black',
    label='Centroids'
)

ax.set_title('3D Customer Segments Visualization', fontsize=16)
ax.set_xlabel('Principal Component 1', fontsize=12)
ax.set_ylabel('Principal Component 2', fontsize=12)
ax.set_zlabel('Principal Component 3', fontsize=12)
ax.legend(title='Segment', title_fontsize=12)
plt.tight_layout()
plt.show()

# 9. Feature importance per cluster
print("\nAnalyzing feature importance for each cluster...")

# Standardize the data again for fair comparison
feature_importance_data = segment_features.copy()
scaler = MinMaxScaler()
feature_importance_scaled = scaler.fit_transform(feature_importance_data)
feature_importance_df = pd.DataFrame(feature_importance_scaled, columns=feature_importance_data.columns)
feature_importance_df['cluster'] = features['cluster']

# Calculate the distance from each cluster's mean to the overall mean for each feature
overall_mean = feature_importance_df.iloc[:, :-1].mean()

cluster_feature_importance = pd.DataFrame()

for cluster in range(final_clusters):
    cluster_mean = feature_importance_df[feature_importance_df['cluster'] == cluster].iloc[:, :-1].mean()
    
    # Calculate difference from overall mean
    difference = cluster_mean - overall_mean
    
    # Add to the dataframe
    cluster_feature_importance[f'Cluster {cluster}: {cluster_names[cluster]}'] = difference

# Visualize feature importance
plt.figure(figsize=(14, len(segment_features.columns) * 0.4))
sns.heatmap(
    cluster_feature_importance.T,
    cmap='coolwarm',
    center=0,
    annot=True,
    fmt='.2f',
    cbar_kws={'label': 'Deviation from Mean (Standardized)'}
)
plt.title('Feature Importance by Cluster', fontsize=16)
plt.tight_layout()
plt.show()

# 10. Create radar charts for cluster profiles
print("\nCreating radar charts for cluster profiles...")

# Function to create a radar chart for cluster profiles
def create_radar_chart(cluster_data, feature_names, cluster_names, title):
    # Number of features
    N = len(feature_names)
    
    # Create angles for each feature
    angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
    
    # Close the polygon
    angles += angles[:1]
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
    
    # Add feature labels
    plt.xticks(angles[:-1], feature_names, size=12)
    
    # Add y-axis labels (0-100%)
    ax.set_rlabel_position(0)
    plt.yticks([0.25, 0.5, 0.75], ["25%", "50%", "75%"], color="grey", size=10)
    plt.ylim(0, 1)
    
    # Plot each cluster
    for i, cluster_id in enumerate(cluster_data.index):
        # Get values for this cluster and close the loop
        values = cluster_data.loc[cluster_id].values.flatten().tolist()
        values += values[:1]
        
        # Plot values
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=cluster_names[cluster_id])
        ax.fill(angles, values, alpha=0.1)
    
    # Add legend
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    
    # Add title
    plt.title(title, size=20, y=1.1)
    
    return fig, ax

# Select features for the radar chart
radar_features = [
    'recency_days', 'frequency', 'monetary_value', 
    'avg_balance', 'loan_count', 'card_count'
]

# Filter to only include available features
radar_features = [f for f in radar_features if f in features.columns]

# Prepare data for radar chart
radar_data = cluster_profile[radar_features].copy()

# Normalize data for radar chart (0-1 scale)
for feature in radar_features:
    max_val = radar_data[feature].max()
    min_val = radar_data[feature].min()
    
    # For recency_days, lower is better, so invert
    if feature == 'recency_days':
        radar_data[feature] = 1 - ((radar_data[feature] - min_val) / (max_val - min_val)) if max_val > min_val else 0.5
    else:
        radar_data[feature] = (radar_data[feature] - min_val) / (max_val - min_val) if max_val > min_val else 0.5

# Create better labels for radar chart
radar_labels = {
    'recency_days': 'Activeness',
    'frequency': 'Transaction Frequency',
    'monetary_value': 'Transaction Value',
    'avg_balance': 'Account Balance',
    'loan_count': 'Loan Usage',
    'card_count': 'Card Usage',
    'permanent_order_count': 'Standing Orders',
    'clv_1yr': 'Customer Value',
    'customer_lifetime_value': 'Customer Value'
}

radar_feature_labels = [radar_labels.get(f, f) for f in radar_features]

# Create radar chart
create_radar_chart(
    radar_data, 
    radar_feature_labels, 
    cluster_names,
    'Banking Customer Segment Profiles'
)
plt.tight_layout()
plt.show()

# 11. Create individual cluster profiles with characteristics and marketing recommendations
print("\nCreating detailed cluster profiles with marketing recommendations...")

# Function to generate marketing recommendations based on cluster characteristics
def generate_recommendations(cluster_id, profile):
    recommendations = []
    
    # Convert Series to dict for easier access
    profile = profile.to_dict()
    
    # Check various characteristics and generate appropriate recommendations
    
    # Recency-based recommendations
    if 'recency_days' in profile:
        if profile['recency_days'] > 60:
            recommendations.append("Re-engagement campaign with special offer to restart activity")
        elif profile['recency_days'] < 30:
            recommendations.append("Reward active usage with loyalty benefits")
    
    # Product diversity recommendations
    product_columns = ['loan_count', 'card_count', 'permanent_order_count']
    available_products = [p for p in product_columns if p in profile]
    
    if available_products:
        total_products = sum([profile.get(p, 0) for p in available_products])
        
        if total_products < 2:
            recommendations.append("Cross-sell additional banking products with bundled benefits")
        
        # Specific product recommendations
        if 'loan_count' in profile and profile['loan_count'] < 0.5:
            recommendations.append("Targeted loan offers based on account history")
        
        if 'card_count' in profile and profile['card_count'] < 0.5:
            recommendations.append("Credit card promotion with first-year fee waiver")
            
        if 'permanent_order_count' in profile and profile['permanent_order_count'] < 1:
            recommendations.append("Promote automatic payment convenience and bill management tools")
    
    # Balance-based recommendations
    if 'avg_balance' in profile:
        if profile['avg_balance'] > cluster_profile['avg_balance'].median():
            recommendations.append("Investment and wealth management product offerings")
        else:
            recommendations.append("Savings account promotions with competitive interest rates")
    
    # CLV-based recommendations
    clv_col = 'clv_1yr' if 'clv_1yr' in profile else 'customer_lifetime_value'
    if clv_col in profile:
        if profile[clv_col] > cluster_profile[clv_col].median():
            recommendations.append("Premium service model with dedicated relationship manager")
            recommendations.append("Exclusive event invitations and premium partnership benefits")
        else:
            recommendations.append("Digital self-service tools for cost-effective account management")
    
    # Transaction-based recommendations
    if 'frequency' in profile:
        if profile['frequency'] > cluster_profile['frequency'].median():
            recommendations.append("Cashback or rewards program for high transaction activity")
        else:
            recommendations.append("Transaction fee discounts to encourage account usage")
    
    return recommendations

# Create individual profiles for each cluster
cluster_detailed_profiles = {}

for cluster_id in range(final_clusters):
    cluster_name = cluster_names[cluster_id]
    cluster_size = cluster_counts[cluster_id]
    cluster_percentage = (cluster_size / len(features) * 100).round(1)
    
    profile = cluster_profile.loc[cluster_id]
    
    # Generate key characteristics
    characteristics = []
    
    # RFM characteristics
    if 'recency_days' in profile:
        if profile['recency_days'] < cluster_profile['recency_days'].median():
            characteristics.append("Recently active customers")
        else:
            characteristics.append("Less recently active customers")
    
    if 'frequency' in profile:
        if profile['frequency'] > cluster_profile['frequency'].mean():
            characteristics.append("High transaction frequency")
        else:
            characteristics.append("Lower transaction frequency")
    
    if 'monetary_value' in profile:
        if profile['monetary_value'] > cluster_profile['monetary_value'].mean():
            characteristics.append("Higher transaction values")
        else:
            characteristics.append("Lower transaction values")
    
    # Balance characteristics
    if 'avg_balance' in profile:
        if profile['avg_balance'] > cluster_profile['avg_balance'].mean():
            characteristics.append("Maintain higher account balances")
        else:
            characteristics.append("Maintain lower account balances")
    
    # Product usage
    product_cols = ['loan_count', 'card_count', 'permanent_order_count']
    available_product_cols = [col for col in product_cols if col in profile]
    
    if available_product_cols:
        for col in available_product_cols:
            if profile[col] > cluster_profile[col].mean():
                product_name = col.replace('_count', '').replace('_', ' ').title()
                characteristics.append(f"Higher {product_name} usage")
    
    # Value characteristics
    clv_col = 'clv_1yr' if 'clv_1yr' in profile else 'customer_lifetime_value'
    if clv_col in profile:
        if profile[clv_col] > cluster_profile[clv_col].mean():
            characteristics.append("Higher lifetime value customers")
        else:
            characteristics.append("Lower lifetime value customers")
    
    # Generate marketing recommendations
    recommendations = generate_recommendations(cluster_id, profile)
    
    # Store the detailed profile
    cluster_detailed_profiles[cluster_id] = {
        'cluster_id': cluster_id,
        'name': cluster_name,
        'size': cluster_size,
        'percentage': cluster_percentage,
        'characteristics': characteristics,
        'recommendations': recommendations,
        'profile': profile
    }

# Display detailed profiles
print("\nDetailed Cluster Profiles:")
for cluster_id, profile in cluster_detailed_profiles.items():
    print(f"\nCluster {cluster_id}: {profile['name']}")
    print(f"Size: {profile['size']} customers ({profile['percentage']}% of total)")
    
    print("\nKey Characteristics:")
    for i, characteristic in enumerate(profile['characteristics'], 1):
        print(f"  {i}. {characteristic}")
    
    print("\nMarketing Recommendations:")
    for i, recommendation in enumerate(profile['recommendations'], 1):
        print(f"  {i}. {recommendation}")
    
    print("\nMetrics:")
    # Print a few key metrics
    key_metrics = ['recency_days', 'frequency', 'monetary_value', 'avg_balance']
    key_metrics = [m for m in key_metrics if m in profile['profile']]
    
    if key_metrics:
        for metric in key_metrics:
            print(f"  {metric}: {profile['profile'][metric]:.2f}")
    
    if clv_col in profile['profile']:
        print(f"  {clv_col}: {profile['profile'][clv_col]:.2f}")
    
    print("-" * 80)

# 12. Save results
print("\nSaving segmentation results...")

# Add cluster names to features dataframe
features['segment_name'] = features['cluster'].map(cluster_names)

# Save the complete segmented customer data
features.to_csv(f'{output_dir}customer_segments.csv', index=False)

# Save cluster profiles
cluster_profile.to_csv(f'{output_dir}cluster_profiles.csv')

# Save marketing recommendations
marketing_recommendations = pd.DataFrame({
    'cluster_id': [p['cluster_id'] for p in cluster_detailed_profiles.values()],
    'segment_name': [p['name'] for p in cluster_detailed_profiles.values()],
    'size': [p['size'] for p in cluster_detailed_profiles.values()],
    'percentage': [p['percentage'] for p in cluster_detailed_profiles.values()],
    'characteristics': [', '.join(p['characteristics']) for p in cluster_detailed_profiles.values()],
    'recommendations': [', '.join(p['recommendations']) for p in cluster_detailed_profiles.values()]
})
marketing_recommendations.to_csv(f'{output_dir}segment_marketing_recommendations.csv', index=False)

print("\nSegmentation analysis complete! Results saved to:")
print(f"- {output_dir}customer_segments.csv")
print(f"- {output_dir}cluster_profiles.csv")
print(f"- {output_dir}segment_marketing_recommendations.csv")

# 13. Prediction function for new customers
print("\nCreating prediction function for new customers...")

# Define a function to predict segment for new customers
def predict_segment(new_customer_data):
    """
    Predict the segment for new customer data
    
    Parameters:
    -----------
    new_customer_data : pandas DataFrame
        DataFrame containing customer features
        
    Returns:
    --------
    pandas DataFrame
        Original data with segment predictions added
    """
    # Ensure we have the required features
    required_features = segment_features.columns
    missing_features = [f for f in required_features if f not in new_customer_data.columns]
    
    if missing_features:
        print(f"Warning: Missing features: {missing_features}")
        # Fill missing features with zeros
        for feature in missing_features:
            new_customer_data[feature] = 0
    
    # Select only the features used in segmentation
    X = new_customer_data[required_features].copy()
    
    # Scale the features
    X_scaled = scaler.transform(X)
    
    # Predict clusters
    cluster_pred = kmeans.predict(X_scaled)
    
    # Add predictions to dataframe
    result = new_customer_data.copy()
    result['cluster'] = cluster_pred
    result['segment_name'] = result['cluster'].map(cluster_names)
    
    return result

# Test the prediction function with a sample customer
# Here we're just using the first few customers from our dataset as an example
sample_customers = features[['client_id'] + list(segment_features.columns)].head(5)

print("\nTesting segment prediction with sample customers:")
predicted_segments = predict_segment(sample_customers)
display(predicted_segments[['client_id', 'cluster', 'segment_name']])

print("\nSegmentation model is ready for deployment!")

# Save the prediction function with joblib
joblib.dump({
    'predict_function': predict_segment,
    'feature_names': list(segment_features.columns),
    'cluster_names': cluster_names,
    'kmeans_model': kmeans,
    'scaler': scaler
}, f'{models_dir}segment_prediction_model.pkl')

print(f"Prediction model saved to {models_dir}segment_prediction_model.pkl")
```
