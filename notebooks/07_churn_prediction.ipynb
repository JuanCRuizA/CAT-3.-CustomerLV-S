```python
# 07_churn_prediction.ipynb
# Banking Customer Churn Prediction Analysis

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from datetime import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, 
    roc_auc_score, confusion_matrix, classification_report, 
    precision_recall_curve, roc_curve
)
from sklearn.inspection import permutation_importance
import shap
import joblib
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline

# For visualization
import matplotlib.ticker as mtick

# Suppress warnings for cleaner output
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')

# Define plot aesthetics
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

# 1. Load the segmented customer data
print("Loading customer data with segments and CLV...")

# Define paths
data_dir = '../data/processed/'
models_dir = '../models/'
output_dir = data_dir
os.makedirs(output_dir, exist_ok=True)

# Load customer features with segments
try:
    features = pd.read_csv(f'{data_dir}customer_segments.csv')
    print(f"Loaded {len(features)} customer records with {features.shape[1]} features")
except FileNotFoundError:
    try:
        # Fallback to pre-segmentation data
        features = pd.read_csv(f'{data_dir}customer_features_with_clv.csv')
        print("Segmentation data not found. Using CLV features.")
    except FileNotFoundError:
        try:
            # Final fallback to original features
            features = pd.read_csv(f'{data_dir}banking_customer_features.csv')
            print("Using original banking features without CLV or segmentation.")
        except FileNotFoundError:
            print("Error: No customer data found. Please run the previous notebooks first.")
            raise

# Display basic information about the dataset
print("\nDataset Overview:")
print(f"Number of customers: {len(features)}")
print(f"Number of features: {features.shape[1]}")
print(f"Feature columns: {', '.join(features.columns[:10])}...")

# Display the first few rows
display(features.head())

# 2. Define churn in the banking context
print("\nDefining churn for banking customers...")

# Explore recency distribution to help define churn threshold
plt.figure(figsize=(10, 6))
if 'recency_days' in features.columns:
    sns.histplot(features['recency_days'], bins=50, kde=True)
    plt.title('Distribution of Customer Recency (Days since Last Transaction)')
    plt.xlabel('Days since Last Transaction')
    plt.ylabel('Number of Customers')
    # Add vertical lines for potential churn thresholds
    plt.axvline(x=90, color='r', linestyle='--', label='90 Days (Typical Churn Definition)')
    plt.axvline(x=60, color='orange', linestyle='--', label='60 Days (Early Warning)')
    plt.axvline(x=180, color='darkred', linestyle='--', label='180 Days (Severe Inactivity)')
    plt.legend()
    plt.show()
    
    # Calculate percentiles to understand distribution
    recency_percentiles = [25, 50, 75, 90, 95]
    percentile_values = np.percentile(features['recency_days'], recency_percentiles)
    percentile_df = pd.DataFrame({'Percentile': recency_percentiles, 'Days': percentile_values})
    print("Recency Percentiles:")
    display(percentile_df)
    
    # Define churn based on recency (using 90 days as standard definition for banking)
    churn_threshold = 90  # Can be adjusted based on business requirements
    features['is_churned'] = (features['recency_days'] > churn_threshold).astype(int)
    
    # Show churn distribution
    churn_distribution = features['is_churned'].value_counts()
    print(f"\nChurn Distribution (>{churn_threshold} days inactivity):")
    print(f"Active Customers: {churn_distribution.get(0, 0)} ({churn_distribution.get(0, 0)/len(features)*100:.1f}%)")
    print(f"Churned Customers: {churn_distribution.get(1, 0)} ({churn_distribution.get(1, 0)/len(features)*100:.1f}%)")
else:
    print("Warning: 'recency_days' column not found. Cannot define churn based on transaction recency.")
    print("Using a dummy churn definition for demonstration purposes.")
    # Create a dummy churn variable if recency isn't available
    features['is_churned'] = np.random.choice([0, 1], size=len(features), p=[0.8, 0.2])

# Visualize churn by segment if segmentation data is available
if 'segment_name' in features.columns:
    plt.figure(figsize=(12, 6))
    segment_churn = features.groupby('segment_name')['is_churned'].mean() * 100
    ax = segment_churn.sort_values(ascending=False).plot(kind='bar', color='darkred')
    plt.title('Churn Rate by Customer Segment')
    plt.xlabel('Customer Segment')
    plt.ylabel('Churn Rate (%)')
    plt.xticks(rotation=45, ha='right')
    # Display percentage on top of bars
    for i, v in enumerate(segment_churn.sort_values(ascending=False)):
        ax.text(i, v + 0.5, f"{v:.1f}%", ha='center')
    plt.tight_layout()
    plt.show()

# 3. Feature selection for churn prediction
print("\nSelecting features for churn prediction...")

# Remove target variable and identifiers from feature set
exclude_cols = ['is_churned', 'client_id', 'cluster', 'segment_name']
X_cols = [col for col in features.columns if col not in exclude_cols]

# Check if we have CLV and segment features - highlight their importance if available
clv_cols = [col for col in X_cols if 'clv' in col.lower() or 'lifetime' in col.lower()]
if clv_cols:
    print(f"Including CLV features in churn prediction: {', '.join(clv_cols)}")

segment_cols = [col for col in X_cols if 'cluster' in col.lower() or 'segment' in col.lower()]
if segment_cols:
    print(f"Including segmentation features in churn prediction: {', '.join(segment_cols)}")

# Define feature groups specific to churn prediction
churn_feature_groups = {
    'Engagement Features': ['recency_days', 'frequency', 'transaction_day_diversity', 
                           'weekend_transaction_ratio', 'transaction_type_diversity'],
    'Transaction Features': ['monetary_value', 'avg_transaction_amount', 'credit_transaction_ratio',
                            'monetary_per_transaction', 'transaction_amount_volatility'],
    'Balance Features': ['avg_balance', 'balance_range', 'balance_volatility', 
                        'min_balance', 'max_balance'],
    'Product Features': ['product_diversity_score', 'product_diversity', 'loan_count', 
                         'card_count', 'permanent_order_count'],
    'CLV Features': clv_cols,
    'Segment Features': segment_cols
}

# Filter to only include available columns
available_churn_features = []
for group, cols in churn_feature_groups.items():
    available_group_cols = [col for col in cols if col in X_cols]
    if available_group_cols:
        print(f"- {group}: {', '.join(available_group_cols)}")
        available_churn_features.extend(available_group_cols)

# Ensure no duplicates
available_churn_features = list(dict.fromkeys(available_churn_features))

# Create our feature set
X = features[available_churn_features].copy()
y = features['is_churned'].copy()

# Check for missing values
missing_values = X.isnull().sum()
if missing_values.sum() > 0:
    print(f"\nFound {missing_values.sum()} missing values in {missing_values[missing_values > 0].shape[0]} columns")
    print(missing_values[missing_values > 0])
    print("Filling missing values...")
    X = X.fillna(0)  # Simple imputation for demonstration

# 4. Exploratory analysis of churn factors
print("\nAnalyzing key churn factors...")

# Function to create feature distribution by churn plots
def plot_feature_by_churn(df, feature, feature_name=None):
    if feature_name is None:
        feature_name = feature.replace('_', ' ').title()
    
    plt.figure(figsize=(12, 5))
    
    # Split into active and churned customers
    active = df[df['is_churned'] == 0][feature]
    churned = df[df['is_churned'] == 1][feature]
    
    # Determine plot type based on unique values
    if df[feature].nunique() <= 5:  # Categorical feature
        # Calculate percentages
        active_counts = df[df['is_churned'] == 0][feature].value_counts(normalize=True) * 100
        churned_counts = df[df['is_churned'] == 1][feature].value_counts(normalize=True) * 100
        
        # Create a DataFrame for plotting
        plot_df = pd.DataFrame({
            'Active': active_counts, 
            'Churned': churned_counts
        }).fillna(0)
        
        # Plot
        ax = plot_df.plot(kind='bar', figsize=(12, 5))
        plt.title(f'Distribution of {feature_name} by Churn Status')
        plt.xlabel(feature_name)
        plt.ylabel('Percentage (%)')
        plt.legend(['Active', 'Churned'])
        
        # Add percentage labels
        for container in ax.containers:
            ax.bar_label(container, fmt='%.1f%%')
    else:  # Continuous feature
        plt.subplot(1, 2, 1)
        sns.histplot(data=df, x=feature, hue='is_churned', bins=30, 
                    element='step', common_norm=False, stat='density')
        plt.title(f'Distribution of {feature_name} by Churn Status')
        plt.xlabel(feature_name)
        plt.ylabel('Density')
        plt.legend(['Active', 'Churned'])
        
        plt.subplot(1, 2, 2)
        sns.boxplot(data=df, x='is_churned', y=feature)
        plt.title(f'Box Plot of {feature_name} by Churn Status')
        plt.xticks([0, 1], ['Active', 'Churned'])
        plt.ylabel(feature_name)
    
    plt.tight_layout()
    plt.show()
    
    # Calculate and return statistical significance
    try:
        from scipy import stats
        stat, p_value = stats.ttest_ind(active, churned, equal_var=False)
        print(f"T-test for {feature_name}: p-value = {p_value:.4f}")
        print(f"Active mean: {active.mean():.2f}, Churned mean: {churned.mean():.2f}")
        print(f"Difference: {(churned.mean() - active.mean()):.2f} ({(churned.mean() - active.mean()) / active.mean() * 100:.1f}%)")
        return p_value
    except:
        print(f"Could not calculate statistical significance for {feature_name}")
        return None

# Analyze most important churn factors
print("\nExploring key churn drivers...")

# Select a few key features to analyze based on availability
key_churn_features = []
for feature_list in [
    ['recency_days', 'frequency', 'monetary_value'],  # Core RFM
    ['avg_balance', 'balance_volatility'],            # Balance behavior
    ['product_diversity_score', 'product_diversity'], # Product adoption
    clv_cols                                          # CLV metrics
]:
    key_churn_features.extend([f for f in feature_list if f in X.columns][:1])  # Take first available

# Show distribution of these key features by churn status
for feature in key_churn_features:
    plot_feature_by_churn(features, feature)

# 5. Feature preparation and scaling
print("\nPreparing features for model training...")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print(f"Training set: {X_train.shape[0]} customers ({sum(y_train)} churned)")
print(f"Testing set: {X_test.shape[0]} customers ({sum(y_test)} churned)")

# Standard scaling of features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Save the scaler for future predictions
joblib.dump(scaler, f'{models_dir}churn_prediction_scaler.pkl')

# 6. Model training and evaluation
print("\nTraining churn prediction models...")

# Define evaluation function to compare models
def evaluate_model(model, X_train, X_test, y_train, y_test, model_name="Model"):
    # Make predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)
    
    # Print results
    print(f"\n{model_name} Performance:")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC: {auc:.4f}")
    
    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Active', 'Churned'],
                yticklabels=['Active', 'Churned'])
    plt.title(f'{model_name} Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()
    
    # Calculate and plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {auc:.3f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} ROC Curve')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Show classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    return {
        'model': model,
        'name': model_name,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'auc': auc,
        'y_pred': y_pred,
        'y_prob': y_prob
    }

# Train models with different approaches

# 1. Logistic Regression (baseline)
print("\nTraining Logistic Regression (baseline model)...")
lr = LogisticRegression(max_iter=1000, class_weight='balanced')
lr.fit(X_train_scaled, y_train)
lr_results = evaluate_model(lr, X_train_scaled, X_test_scaled, y_train, y_test, "Logistic Regression")

# 2. Random Forest with class weighting
print("\nTraining Random Forest...")
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    class_weight='balanced',
    random_state=42
)
rf.fit(X_train_scaled, y_train)
rf_results = evaluate_model(rf, X_train_scaled, X_test_scaled, y_train, y_test, "Random Forest")

# 3. Gradient Boosting with SMOTE (for handling imbalance)
print("\nTraining Gradient Boosting with SMOTE...")
# Define SMOTE pipeline
smote_pipeline = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ('classifier', GradientBoostingClassifier(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=5,
        random_state=42
    ))
])
smote_pipeline.fit(X_train_scaled, y_train)
gb_results = evaluate_model(smote_pipeline, X_train_scaled, X_test_scaled, y_train, y_test, "Gradient Boosting with SMOTE")

# Select the best model based on F1 score (balances precision and recall)
models = [lr_results, rf_results, gb_results]
best_model = max(models, key=lambda x: x['f1'])
print(f"\nBest model: {best_model['name']} with F1 Score: {best_model['f1']:.4f}")

# 7. Feature importance analysis
print("\nAnalyzing feature importance for churn prediction...")

# 7.1 Basic feature importance
if hasattr(best_model['model'], 'feature_importances_'):
    # For tree-based models
    importances = best_model['model'].feature_importances_
    if isinstance(best_model['model'], Pipeline):
        # If it's a pipeline, get the classifier's feature importances
        importances = best_model['model'].named_steps['classifier'].feature_importances_
    
    # Sort features by importance
    feature_importance = pd.DataFrame({
        'Feature': X.columns,
        'Importance': importances
    }).sort_values('Importance', ascending=False)
    
    # Plot feature importance
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=feature_importance.head(15))
    plt.title(f'Top 15 Features for Churn Prediction ({best_model["name"]})')
    plt.tight_layout()
    plt.show()
    
    # Display top features
    print("\nTop 10 Features for Churn Prediction:")
    display(feature_importance.head(10))
elif hasattr(best_model['model'], 'coef_'):
    # For linear models
    coefficients = best_model['model'].coef_[0]
    feature_importance = pd.DataFrame({
        'Feature': X.columns,
        'Coefficient': coefficients
    }).sort_values('Coefficient', key=abs, ascending=False)
    
    # Plot feature importance
    plt.figure(figsize=(12, 8))
    colors = ['red' if x < 0 else 'green' for x in feature_importance.head(15)['Coefficient']]
    sns.barplot(x='Coefficient', y='Feature', data=feature_importance.head(15), palette=colors)
    plt.title(f'Top 15 Features for Churn Prediction ({best_model["name"]})')
    plt.axvline(x=0, color='k', linestyle='--')
    plt.tight_layout()
    plt.show()
    
    # Display top features
    print("\nTop 10 Features for Churn Prediction:")
    display(feature_importance.head(10))
else:
    # If model doesn't provide direct feature importance
    print("Computing permutation feature importance (this may take a while)...")
    perm_importance = permutation_importance(
        best_model['model'], X_test_scaled, y_test, 
        n_repeats=10, random_state=42
    )
    
    # Sort features by importance
    feature_importance = pd.DataFrame({
        'Feature': X.columns,
        'Importance': perm_importance.importances_mean
    }).sort_values('Importance', ascending=False)
    
    # Plot feature importance
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Importance', y='Feature', data=feature_importance.head(15))
    plt.title(f'Top 15 Features for Churn Prediction (Permutation Importance)')
    plt.tight_layout()
    plt.show()
    
    # Display top features
    print("\nTop 10 Features for Churn Prediction:")
    display(feature_importance.head(10))

# 7.2 SHAP analysis for deeper insights (if possible)
try:
    print("\nComputing SHAP values for deeper feature insights...")
    
    if isinstance(best_model['model'], Pipeline):
        # If using SMOTE pipeline, get the classifier
        estimator = best_model['model'].named_steps['classifier']
    else:
        estimator = best_model['model']
    
    # Create explainer based on model type
    if hasattr(estimator, 'predict_proba'):
        # For tree-based models
        if hasattr(estimator, 'estimators_'):
            explainer = shap.TreeExplainer(estimator)
            shap_values = explainer.shap_values(X_test_scaled)
            
            # For Random Forest/Gradient Boosting, SHAP returns a list of arrays
            if isinstance(shap_values, list):
                shap_values = shap_values[1]  # Use values for churned class
        else:
            # For linear models
            explainer = shap.LinearExplainer(estimator, X_train_scaled)
            shap_values = explainer.shap_values(X_test_scaled)
        
        # Plot SHAP summary
        plt.figure(figsize=(12, 10))
        shap.summary_plot(
            shap_values, 
            X_test_scaled, 
            feature_names=X.columns, 
            show=False
        )
        plt.title('SHAP Feature Importance for Churn Prediction')
        plt.tight_layout()
        plt.show()
        
        # Show SHAP decision plot for a sample of customers
        sample_indices = np.random.choice(len(X_test), size=min(5, len(X_test)), replace=False)
        
        plt.figure(figsize=(15, 10))
        shap.decision_plot(
            explainer.expected_value if hasattr(explainer, 'expected_value') else 0, 
            shap_values[sample_indices], 
            X_test_scaled[sample_indices],
            feature_names=list(X.columns),
            show=False
        )
        plt.title('SHAP Decision Plot for Sample Customers')
        plt.tight_layout()
        plt.show()
except Exception as e:
    print(f"SHAP analysis error: {e}")
    print("Skipping detailed SHAP analysis")

# 8. Optimizing probability threshold for business impact
print("\nOptimizing churn prediction threshold for business impact...")

# Define cost-benefit parameters (adjust based on business reality)
cost_per_retention_campaign = 100  # Cost to engage with a customer to prevent churn
value_of_retained_customer = 1000  # Value (CLV) of retaining a customer who would have churned

if best_model['name'] == 'Logistic Regression':
    # For interpretability, we'll also calculate odds ratios for Logistic Regression
    print("\nLogistic Regression Odds Ratios:")
    odds_ratios = pd.DataFrame({
        'Feature': X.columns,
        'Coefficient': best_model['model'].coef_[0],
        'Odds Ratio': np.exp(best_model['model'].coef_[0])
    }).sort_values('Odds Ratio', ascending=False)
    display(odds_ratios)

# Calculate precision, recall, and profit for different thresholds
thresholds = np.arange(0.05, 0.95, 0.05)
threshold_metrics = []

for threshold in thresholds:
    y_pred_t = (best_model['y_prob'] >= threshold).astype(int)
    
    # Calculate basic metrics
    precision = precision_score(y_test, y_pred_t)
    recall = recall_score(y_test, y_pred_t)
    f1 = f1_score(y_test, y_pred_t)
    
    # Calculate business metrics
    true_positives = ((y_test == 1) & (y_pred_t == 1)).sum()
    false_positives = ((y_test == 0) & (y_pred_t == 1)).sum()
    
    # Campaign costs
    campaign_cost = (true_positives + false_positives) * cost_per_retention_campaign
    
    # Retention benefit (assuming campaigns have a 50% success rate)
    retention_success_rate = 0.5
    retention_benefit = true_positives * retention_success_rate * value_of_retained_customer
    
    # Net profit
    net_profit = retention_benefit - campaign_cost
    
    # Add to results
    threshold_metrics.append({
        'Threshold': threshold,
        'Precision': precision,
        'Recall': recall,
        'F1': f1,
        'Campaigns': true_positives + false_positives,
        'True Positives': true_positives,
        'False Positives': false_positives,
        'Campaign Cost': campaign_cost,
        'Retention Benefit': retention_benefit,
        'Net Profit': net_profit
    })

# Convert to dataframe
threshold_df = pd.DataFrame(threshold_metrics)

# Plot the results
plt.figure(figsize=(15, 10))

# Metric curves
plt.subplot(2, 2, 1)
plt.plot(threshold_df['Threshold'], threshold_df['Precision'], 'b-', linewidth=2, label='Precision')
plt.plot(threshold_df['Threshold'], threshold_df['Recall'], 'r-', linewidth=2, label='Recall')
plt.plot(threshold_df['Threshold'], threshold_df['F1'], 'g-', linewidth=2, label='F1 Score')
plt.title('Performance Metrics vs. Threshold')
plt.xlabel('Threshold')
plt.ylabel('Metric Value')
plt.grid(True, alpha=0.3)
plt.legend()

# Profit curve
plt.subplot(2, 2, 2)
plt.plot(threshold_df['Threshold'], threshold_df['Net Profit'], 'purple', linewidth=2)
plt.title('Net Profit vs. Threshold')
plt.xlabel('Threshold')
plt.ylabel('Net Profit')
plt.grid(True, alpha=0.3)

# Optimal threshold
optimal_threshold = threshold_df.loc[threshold_df['Net Profit'].idxmax(), 'Threshold']
plt.axvline(x=optimal_threshold, color='red', linestyle='--', 
            label=f'Optimal Threshold = {optimal_threshold:.2f}')
plt.legend()

# Cost-benefit components
plt.subplot(2, 2, 3)
plt.plot(threshold_df['Threshold'], threshold_df['Campaign Cost'], 'r-', linewidth=2, label='Campaign Cost')
plt.plot(threshold_df['Threshold'], threshold_df['Retention Benefit'], 'g-', linewidth=2, label='Retention Benefit')
plt.title('Cost and Benefit Components')
plt.xlabel('Threshold')
plt.ylabel('Value')
plt.grid(True, alpha=0.3)
plt.legend()

# Campaign volume vs threshold
plt.subplot(2, 2, 4)
bar_width = 0.35
index = np.arange(len(thresholds))
plt.bar(index, threshold_df['True Positives'], bar_width, label='True Positives (Actual Churners)', color='g')
plt.bar(index, threshold_df['False Positives'], bar_width, bottom=threshold_df['True Positives'],
        label='False Positives (Non-churners)', color='r')
plt.xlabel('Threshold')
plt.ylabel('Number of Customers')
plt.title('Campaign Volume by Threshold')
plt.xticks(index, [f'{t:.2f}' for t in thresholds], rotation=90)
plt.legend()

plt.tight_layout()
plt.show()

# Display optimal threshold and results
print(f"\nOptimal threshold for maximizing profit: {optimal_threshold:.2f}")
optimal_results = threshold_df[threshold_df['Threshold'] == optimal_threshold].iloc[0]
print(f"Precision at optimal threshold: {optimal_results['Precision']:.3f}")
print(f"Recall at optimal threshold: {optimal_results['Recall']:.3f}")
print(f"Estimated net profit: ${optimal_results['Net Profit']:,.2f}")
print(f"Number of retention campaigns: {int(optimal_results['Campaigns'])}")
print(f"True churners targeted: {int(optimal_results['True Positives'])} out of {sum(y_test)} ({100*optimal_results['True Positives']/sum(y_test):.1f}%)")

# 9. Identify high-value customers at risk of churn
print("\nIdentifying high-value customers at risk of churn...")

# Apply optimal threshold to get churn predictions
y_pred_optimal = (best_model['y_prob'] >= optimal_threshold).astype(int)

# Create a prediction dataframe
prediction_df = pd.DataFrame({
    'client_id': features.iloc[y_test.index]['client_id'],
    'actual_churn': y_test,
    'predicted_churn': y_pred_optimal,
    'churn_probability': best_model['y_prob']
})

# Add additional information from original data
for col in ['recency_days', 'frequency', 'monetary_value', 'avg_balance']:
    if col in features.columns:
        prediction_df[col] = features.iloc[y_test.index][col].values

# Add CLV information if available
clv_col = None
for col in ['clv_1yr', 'customer_lifetime_value']:
    if col in features.columns:
        clv_col = col
        prediction_df[col] = features.iloc[y_test.index][col].values
        break

# Add segment information if available
if 'segment_name' in features.columns:
    prediction_df['segment'] = features.iloc[y_test.index]['segment_name'].values

# Identify high-value at-risk customers
if clv_col:
    # Define threshold for "high value" (e.g., top 25%)
    high_value_threshold = prediction_df[clv_col].quantile(0.75)
    
    high_value_at_risk = prediction_df[
        (prediction_df['churn_probability'] >= optimal_threshold) & 
        (prediction_df[clv_col] >= high_value_threshold)
    ].sort_values(clv_col, ascending=False)
    
    print(f"\nIdentified {len(high_value_at_risk)} high-value customers at risk of churn")
    print(f"High value threshold: {high_value_threshold:.2f} ({clv_col})")
    
    # Display top 10 high-value at-risk customers
    print("\nTop 10 high-value customers at risk of churn:")
    display(high_value_at_risk.head(10))
    
    # Save high-value at-risk customers to CSV
    high_value_at_risk.to_csv(f'{output_dir}high_value_churn_risk.csv', index=False)
    print(f"Saved {len(high_value_at_risk)} high-value at-risk customers to {output_dir}high_value_churn_risk.csv")
    
    # Visualize high-value at-risk customers by segment (if available)
    if 'segment' in high_value_at_risk.columns:
        if len(high_value_at_risk) > 0:  # Check if we have any high-value customers at risk
            plt.figure(figsize=(12, 6))
            segment_counts = high_value_at_risk['segment'].value_counts()
            
            # Only create plot if we have segments with values
            if len(segment_counts) > 0:
                ax = segment_counts.plot(kind='bar', color='darkred')
                plt.title('High-Value At-Risk Customers by Segment')
                plt.xlabel('Customer Segment')
                plt.ylabel('Number of Customers')
                plt.xticks(rotation=45, ha='right')
                # Display count on top of bars
                for i, v in enumerate(segment_counts):
                    ax.text(i, v + 0.1, str(v), ha='center')
                plt.tight_layout()
                plt.show()
            else:
                print("No segment data available for high-value at-risk customers")
        else:
            print("No high-value customers identified as being at risk of churn")
else:
    print("CLV data not available - cannot identify high-value customers at risk")
    
    # Just save the predictions
    prediction_df.to_csv(f'{output_dir}churn_predictions.csv', index=False)
    print(f"Saved churn predictions to {output_dir}churn_predictions.csv")

# 10. Save the best model
print("\nSaving the best churn prediction model...")

# Save model and related components
joblib.dump(best_model['model'], f'{models_dir}churn_model.pkl')
joblib.dump({
    'model_type': best_model['name'],
    'features': list(X.columns),
    'optimal_threshold': optimal_threshold,
    'model_metrics': {
        'accuracy': best_model['accuracy'],
        'precision': best_model['precision'],
        'recall': best_model['recall'],
        'f1': best_model['f1'],
        'auc': best_model['auc']
    },
    'feature_importance': feature_importance.to_dict(),
    'creation_date': datetime.now().strftime('%Y-%m-%d')
}, f'{models_dir}churn_model_metadata.pkl')

print(f"Model and metadata saved to {models_dir}")

# 11. Create prediction function for future use
def predict_churn(customer_data, threshold=None):
    """
    Predict churn probability for new customer data
    
    Parameters:
    -----------
    customer_data : pandas DataFrame
        DataFrame containing customer features
    threshold : float, optional
        Custom threshold for churn classification
        
    Returns:
    --------
    pandas DataFrame
        Original data with churn predictions and probabilities added
    """
    if threshold is None:
        threshold = optimal_threshold
    
    # Load the model and scaler
    model = joblib.load(f'{models_dir}churn_model.pkl')
    scaler = joblib.load(f'{models_dir}churn_prediction_scaler.pkl')
    
    # Ensure we have the required features
    required_features = X.columns
    missing_features = [f for f in required_features if f not in customer_data.columns]
    
    if missing_features:
        print(f"Warning: Missing features: {missing_features}")
        # Fill missing features with zeros
        for feature in missing_features:
            customer_data[feature] = 0
    
    # Select only the features used in the model
    X_new = customer_data[required_features].copy()
    
    # Scale the features
    X_new_scaled = scaler.transform(X_new)
    
    # Predict probabilities
    churn_probabilities = model.predict_proba(X_new_scaled)[:, 1]
    
    # Determine churn based on optimal threshold
    churn_predictions = (churn_probabilities >= threshold).astype(int)
    
    # Add predictions to dataframe
    result = customer_data.copy()
    result['churn_probability'] = churn_probabilities
    result['is_churned'] = churn_predictions
    
    return result

# Save the prediction function
joblib.dump(predict_churn, f'{models_dir}predict_churn_function.pkl')

# Create banking-specific churn levels for easier interpretation
def assign_churn_risk_level(probability):
    if probability < 0.2:
        return "Very Low"
    elif probability < 0.4:
        return "Low"
    elif probability < 0.6:
        return "Medium"
    elif probability < 0.8:
        return "High"
    else:
        return "Very High"

# 12. Generate a sample retention strategy report
print("\nGenerating retention strategy recommendations...")

# Identify at-risk segments and characteristics
if 'segment' in prediction_df.columns:
    # Calculate churn risk by segment
    segment_risk = prediction_df.groupby('segment')['churn_probability'].mean().sort_values(ascending=False)
    
    # For the highest risk segment, analyze key factors
    highest_risk_segment = segment_risk.index[0]
    segment_customers = features[features['segment_name'] == highest_risk_segment]
    
    print(f"\nRetention Strategy for High-Risk Segment: {highest_risk_segment}")
    print(f"Average Churn Probability: {segment_risk[highest_risk_segment]:.2f}")
    
    # List key characteristics of this segment
    print("\nKey Characteristics:")
    for feature in key_churn_features:
        if feature in segment_customers.columns:
            segment_avg = segment_customers[feature].mean()
            overall_avg = features[feature].mean()
            diff_pct = (segment_avg - overall_avg) / overall_avg * 100
            print(f"- {feature}: {segment_avg:.2f} ({diff_pct:+.1f}% vs. overall average)")
    
    # Generate tailored recommendations based on segment characteristics
    print("\nRecommended Retention Strategies:")
    if 'recency_days' in features.columns and segment_customers['recency_days'].mean() > features['recency_days'].mean():
        print("1. Re-engagement campaign with personalized offers")
        print("   - Send targeted communications highlighting new services")
        print("   - Offer incentives for next transaction")
    
    if 'avg_balance' in features.columns and segment_customers['avg_balance'].mean() < features['avg_balance'].mean():
        print("2. Savings account promotion")
        print("   - Offer competitive interest rates for new deposits")
        print("   - Provide financial planning tools and education")
    
    if 'product_diversity_score' in features.columns and segment_customers['product_diversity_score'].mean() < features['product_diversity_score'].mean():
        print("3. Cross-selling campaign")
        print("   - Bundle complementary products with special pricing")
        print("   - Highlight benefits of additional banking products")
    
    if 'frequency' in features.columns and segment_customers['frequency'].mean() < features['frequency'].mean():
        print("4. Usage incentive program")
        print("   - Reward program for regular account activity")
        print("   - Reduce or waive fees for increased usage")
    
    # General recommendations
    print("\nGeneral Retention Recommendations:")
    print("1. Enhanced customer service for high-value customers")
    print("2. Personalized digital banking experience")
    print("3. Proactive outreach before key churn indicators appear")
    print("4. Competitive analysis to identify stronger competitor offerings")
    print("5. Feedback collection to identify pain points and areas for improvement")

# 13. Conclusion and next steps
print("\n=== Churn Prediction Model Summary ===")
print(f"Model Type: {best_model['name']}")
print(f"Optimal Threshold: {optimal_threshold:.2f}")
print(f"AUC Score: {best_model['auc']:.4f}")
print(f"Precision at Optimal Threshold: {optimal_results['Precision']:.4f}")
print(f"Recall at Optimal Threshold: {optimal_results['Recall']:.4f}")
print(f"Top Churn Indicator: {feature_importance.iloc[0]['Feature']}")

print("\nNext Steps:")
print("1. Implement retention campaign for high-value at-risk customers")
print("2. Create automated early warning system for emerging churn risk")
print("3. Develop segment-specific retention strategies")
print("4. Monitor churn model performance and retrain as needed")
print("5. Connect churn predictions with CRM and marketing automation systems")
```
