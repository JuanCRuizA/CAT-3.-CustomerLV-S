DATABASE SCHEMA INFORMATION


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyodbc

conn = pyodbc.connect('Driver={SQL Server};'
                     'Server=JUANCARLOSRUIZA;'
                     'Database=CzechBankingAnalysis;'
                     'Trusted_Connection=yes;')
```

```python
# Query to retrieve complete database schema information
schema_query = """
SELECT 
    t.TABLE_NAME,
    c.COLUMN_NAME,
    c.ORDINAL_POSITION as Column_Order,
    c.DATA_TYPE,
    c.CHARACTER_MAXIMUM_LENGTH as Max_Length,
    c.NUMERIC_PRECISION as Numeric_Precision,
    c.NUMERIC_SCALE as Numeric_Scale,
    c.IS_NULLABLE as Is_Nullable
FROM 
    INFORMATION_SCHEMA.TABLES t
INNER JOIN 
    INFORMATION_SCHEMA.COLUMNS c 
    ON t.TABLE_NAME = c.TABLE_NAME
WHERE 
    t.TABLE_TYPE = 'BASE TABLE'
    AND t.TABLE_CATALOG = 'CzechBankingAnalysis'  -- Your database name
ORDER BY 
    t.TABLE_NAME, 
    c.ORDINAL_POSITION
"""

# Execute the query and display results
schema_info = pd.read_sql(schema_query, conn)
display(schema_info)
```

DATA TYPE FIELD SAMPLES

```python

# 1. Account.acc_date samples
print("=== Account.acc_date samples ===")
acc_date_query = """
SELECT TOP 5 account_id, acc_date,
       -- Try different ways of formatting the date to understand its stored format
       CONVERT(VARCHAR, acc_date, 120) AS iso_date,  -- YYYY-MM-DD HH:MI:SS
       YEAR(acc_date) AS year,
       MONTH(acc_date) AS month,
       DAY(acc_date) AS day
FROM Account
ORDER BY account_id
"""
acc_date_samples = pd.read_sql(acc_date_query, conn)
display(acc_date_samples)
```

```python
# 2. Client.birth_number samples (not actually a date but might be date-related)
print("\n=== Client.birth_number samples ===")
birth_number_query = """
SELECT TOP 5 client_id, birth_number
FROM Client
ORDER BY client_id
"""
birth_number_samples = pd.read_sql(birth_number_query, conn)
display(birth_number_samples)
```

```python
# 3. CreditCard.issued samples
print("\n=== CreditCard.issued samples ===")
card_issued_query = """
SELECT TOP 5 card_id, issued,
       CONVERT(VARCHAR, issued, 120) AS iso_date,
       YEAR(issued) AS year,
       MONTH(issued) AS month,
       DAY(issued) AS day
FROM CreditCard
ORDER BY card_id
"""
card_issued_samples = pd.read_sql(card_issued_query, conn)
display(card_issued_samples)
```
